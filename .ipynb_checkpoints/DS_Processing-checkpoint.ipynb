```python
def add_student(name,gender,email,course,major):
    mydb = psycopg2.connect(database=config["database"], user=config["user"], password=config["password"], host=config["host"], port=config["port"], options=f'-c search_path=biabsence')
    mycursor = mydb.cursor()
    mycursor.execute(f"SELECT * FROM students where email ='{email}'")
    myresult = mycursor.fetchall()
    if len(myresult) != 0:
        return myresult[0][0]
    if len(myresult) == 0:
        sql = "INSERT INTO students (name,gender,email,course,major) VALUES (%s,%s,%s,%s,%s)"
        val = (name,gender,email,course,major)
        mycursor.execute(sql, val)
        mydb.commit()
        # print(mycursor.rowcount, "record inserted.")
        mycursor.execute(f"SELECT * FROM students where email ='{email}'")
        myresult = mycursor.fetchall()
        return myresult[0][0]
```

```python
df_ds = pd.read_excel("/home/anas/notebook/UseCase/rawdata_lists/Liste des 2èmes années DS.xlsx",engine="openpyxl")
name=[]
gender=[]
email=[]
course="DWH"
major="DS"


for index, row in df_ds.iterrows():
    #Traitement des emails
    Nom_=row['Nom']
    prenom_ = Nom_.split(" ", 1)[0]
    nom_ = Nom_.split(" ", 1)[1]
    
    l = nom_.split(' ')
    n = ""
    for i in l :
        i = ''.join(filter(str.isalnum, i))
        n += i
    firstchar = prenom_.split('-')
    first = ''
    for j in firstchar :
        first += j[0]
        
    email_=f"{first.lower()}{n.lower()}@insea.ac.ma"
    email.append(email_)
    name.append(Nom_)
    
    #Traitement du sexe
    gender_=""
    s=["amina", "asma", "chaimaa", "chaimae", "hajar", "ibtihal", "imane", "kawtar", "nohaila", "soukaina", "yasmine"]
    if prenom_.lower() in s:
        gender_= "f"
    else:   
        gender_="m"
    gender.append(gender_)
    
    
df_ds["name"]=name
df_ds["gender"]=gender
df_ds["email"]=email
df_ds["course"]=course
df_ds["major"]=major
df_ds = df_ds.drop('Nom', 1)
df_ds = df_ds.dropna(1, how='all')
df_ds
```
