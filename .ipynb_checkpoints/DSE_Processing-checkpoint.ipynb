```python
def add_student(name,gender,email,course,major):
    mydb = psycopg2.connect(database=config["database"], user=config["user"], password=config["password"], host=config["host"], port=config["port"], options=f'-c search_path=biabsence')
    mycursor = mydb.cursor()
    mycursor.execute(f"SELECT * FROM students where email ='{email}'")
    myresult = mycursor.fetchall()
    if len(myresult) != 0:
        return myresult[0][0]
    if len(myresult) == 0:
        sql = "INSERT INTO students (name,gender,email,course,major) VALUES (%s,%s,%s,%s,%s)"
        val = (name,gender,email,course,major)
        mycursor.execute(sql, val)
        mydb.commit()
        # print(mycursor.rowcount, "record inserted.")
        mycursor.execute(f"SELECT * FROM students where email ='{email}'")
        myresult = mycursor.fetchall()
        return myresult[0][0]
```

```python
df_dse = pd.read_excel("/home/anas/notebook/UseCase/rawdata_lists/Liste des 2èmes années DSE.xlsx",engine="openpyxl")
name=[]
gender=[]
email=[]
course="DWH"
major="DSE"


for index, row in df_dse.iterrows():
    #Traitement des emails
    Nom_=row['Nom']
    prenom_ = Nom_.split(" ", 1)[0]
    nom_ = Nom_.split(" ", 1)[1]
    
    l = nom_.split(' ')
    n = ""
    for i in l :
        i = ''.join(filter(str.isalnum, i))
        n += i
    firstchar = prenom_.split('-')
    first = ''
    for j in firstchar :
        first += j[0]
        
    email_=f"{first.lower()}{n.lower()}@insea.ac.ma"
    email.append(email_)
    name.append(Nom_)
    
    #Traitement du sexe
    gender_=""
    s=["fathiya", "sabah", "chaimaa", "chaymae", "hibatallah", "hajar", "niima", "fadwa", "soukayna", "aicha", "soukaina", "hafsa", "hala", "khaoula", "amal", "zineb"]
    if prenom_.lower() in s:
        gender_= "f"
    else:   
        gender_="m"
    gender.append(gender_)
    
df_dse["name"]=name
df_dse["gender"]=gender
df_dse["email"]=email
df_dse["course"]=course
df_dse["major"]=major
df_dse = df_dse.drop('Nom', 1)
df_dse = df_dse.dropna(1, how='all')
df_dse
```
